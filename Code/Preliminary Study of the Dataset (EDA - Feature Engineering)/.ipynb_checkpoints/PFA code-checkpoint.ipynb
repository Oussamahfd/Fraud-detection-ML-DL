```python
def type_de_donnee(serie):
    if pd.api.types.is_numeric_dtype(serie):
        return "Quantitatif"
    else:
        return "Catégoriel"

def echelle_de_mesure(serie):
    if isinstance(serie.dtype, pd.CategoricalDtype) or pd.api.types.is_object_dtype(serie):
        return "Nominale"  # Par défaut pour les données catégorielles, peut être ajusté
    elif pd.api.types.is_integer_dtype(serie):
        return "Ratio"
    elif pd.api.types.is_float_dtype(serie):
        return "Intervalle"
    else:
        return "Inconnu"

def statistiques_descriptives(serie):
    if type_de_donnee(serie) == "Quantitatif":
        return {
            "Moyenne": serie.mean(),
            "Médiane": serie.median(),
            "Mode": serie.mode()[0] if not serie.mode().empty else np.nan,
            "Écart-type": serie.std(),
            "Minimum": serie.min(),
            "Maximum": serie.max()
        }
    else:
        return serie.value_counts(normalize=True).to_dict()

def resume_colonne(df):
    resume = {}
    for colonne in df.columns:
        serie = df[colonne]
        resume[colonne] = {
            "Type de données": type_de_donnee(serie),
            "Échelle de mesure": echelle_de_mesure(serie),
            "Statistiques descriptives": statistiques_descriptives(serie)
        }
    return resume
```

```python
resume = resume_colonne(df)

# Limiter la quantité de données affichées
resume_df = pd.DataFrame(resume).T
resume_df.to_csv('resume_colonnes.csv')

print("Les résultats ont été exportés vers 'resume_colonnes.csv'.")
```

```python
def plotDistPlot(columns):
    fig, ax = plt.subplots(len(columns) // 3 + 1, 3, figsize=(15, 15))
    for i, col in enumerate(columns):
        sns.histplot(df.loc[:, col], stat="percent", kde=True, ax=ax[i // 3, i % 3])
        ax[i // 3, i % 3].set_title(col)
        ax[i // 3, i % 3].legend(labels=['Normal', 'Actual'])
    plt.tight_layout()

plotDistPlot(variables_numeriques)
plt.show()
```

```python
# Créer une figure avec un sous-graphique pour chaque variable numérique
plt.figure(figsize=(15, 10))  # Ajustez la taille selon le nombre de variables

for i, colonne in enumerate(variables_numeriques, 1):
    plt.subplot(len(variables_numeriques), 1, i)
    sns.boxplot(data=df, x=colonne)
    plt.title(f'Boxplot de {colonne}')

plt.tight_layout()
plt.show()
```

```python
# Créer un dictionnaire pour stocker les valeurs aberrantes
valeurs_aberrantes = {}

# Identifier les valeurs aberrantes pour chaque variable numérique
for colonne in variables_numeriques:
    Q1 = df[colonne].quantile(0.25)
    Q3 = df[colonne].quantile(0.75)
    IQR = Q3 - Q1
    
    seuil_inf = Q1 - 1.5 * IQR
    seuil_sup = Q3 + 1.5 * IQR
    
    valeurs_aberrantes[colonne] = df[(df[colonne] < seuil_inf) | (df[colonne] > seuil_sup)]

# Afficher les valeurs aberrantes pour chaque variable numérique
for colonne, df_aberrantes in valeurs_aberrantes.items():
    print(f"Valeurs aberrantes pour {colonne}:")
    print(df_aberrantes[colonne].head())  # Affiche les premières valeurs aberrantes
    print()
```

<H1>Valeurs aberrantes</H1>

```python
from scipy import stats

# Calculez les Z-scores
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))
df_z_scores = pd.DataFrame(z_scores, columns=df.select_dtypes(include=['float64', 'int64']).columns)

# Définissez un seuil pour les valeurs aberrantes
threshold = 3
outliers = (df_z_scores > threshold).any(axis=1)

# Affichez les lignes avec des valeurs aberrantes
df_outliers = df[outliers]
```

```python
from sklearn.ensemble import IsolationForest

# Appliquez Isolation Forest
iso_forest = IsolationForest(contamination=0.01)  # Ajustez le paramètre contamination en fonction de vos besoins
outliers = iso_forest.fit_predict(df.select_dtypes(include=['float64', 'int64']))

# -1 indique les valeurs aberrantes
df['outliers'] = outliers
df_outliers_ml = df[df['outliers'] == -1]
df_outliers_ml
```
