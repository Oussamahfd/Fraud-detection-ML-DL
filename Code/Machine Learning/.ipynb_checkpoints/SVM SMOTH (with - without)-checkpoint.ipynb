# <span style="color:black;">Importation des Bibliothèques et Préparation des Outils pour l'Analyse
</span>


```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
```

# <span style="color:black;">Préparation et Prétraitement des Données pour la Détection de Fraude
</span>


## <span style="color:black;">Chargement des datasets</span>


```python
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

## <span style="color:black;">Suppression des colonnes inutiles</span>


```python
df_train.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
df_test.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
```

## <span style="color:black;">Conversion et Encodage des Colonnes</span>


```python
# Conversion des colonnes de dates en timestamps
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_train['dob'] = pd.to_datetime(df_train['dob']).apply(lambda x: x.timestamp())
```

```python
# Initialize the LabelEncoder
encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = [ 'category', 'city', 'state','dob']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_train[col] = encoder.fit_transform(df_train[col])
```

```python
# List of columns to encode
columns_to_encode = ['category', 'city', 'state','dob']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_test[col] = encoder.fit_transform(df_test[col])

# The rest of your numeric columns should already be in the correct format

df_test ['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test ['trans_date_trans_time'] = df_test ['trans_date_trans_time'].apply(lambda x: x.timestamp())
```

## <span style="color:black;">Séparation des features et de la target</span>


```python
X = df_train .drop(columns=['is_fraud'])
Y = df_train ['is_fraud']
```

## <span style="color:black;">Normalisation des features

</span>


```python
scaler = StandardScaler()
X= scaler.fit_transform(X)
```


## <span style="color:black;"> Séparation des données en train et test

</span>



```python
X_train, _ , Y_train, _ = train_test_split(X, Y, random_state=42)
```

## <span style="color:black;">Échantillonnage d'un sous-ensemble de données pour l'entraînement du modèle

</span>


```python
X_train_subset, _, Y_train_subset, _ = train_test_split(X, Y, train_size=100000, random_state=42)
```

## <span style="color:black;">Afficher la taille de X_train_subset et Y_train_subset
</span>


```python
print("Taille de X_train_subset :", X_train_subset.shape)

print("Taille de Y_train_subset :", Y_train_subset.shape)
```

## <span style="color:black;">Séparation des caractéristiques (X) et des étiquettes (Y) dans le jeu de données de test


</span>


```python
X_test_initial = df_test .drop(columns=['is_fraud'])
Y_test_initial = df_test ['is_fraud']
```

## <span style="color:black;">Normalisation des caractéristiques du jeu de données de test à l'aide de StandardScaler


</span>


```python
scaler = StandardScaler()
X_test_initial= scaler.fit_transform(X_test_initial)
```

## <span style="color:black;"> Division du jeu de données de test normalisé en ensembles de test et de validation


</span>


```python
X_test, _ , Y_test, _ = train_test_split(X_test_initial, Y_test_initial, random_state=42)
```

# <span style="color:black;">Entraînement du Modèle de SVM [Sans / Avec] SMOTE</span>


## <span style="color:black;">Sans SMOTE</span>

```python
# Initialiser le modèle SVM avec un noyau linéaire pour une exécution plus rapide
svm_model_without_smote = SVC(kernel='linear', random_state=42)

# Entraîner le modèle sur les données d'entraînement
svm_model_without_smote.fit(X_train_subset, Y_train_subset)
```

```python
# Prédire sur les données de test
y_pred_without_smote = svm_model_without_smote.predict(X_test)
```

#### <span style="color:black;">Évaluation des Performances du Modèle d'arbre de décision</span>


```python
# Calculer l'accuracy
accuracy_without_smote = accuracy_score(Y_test, y_pred_without_smote)

# Obtenir le rapport de classification sous forme de dictionnaire
report_without_smote = classification_report(Y_test, y_pred_without_smote, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_without_smote = pd.DataFrame(report_without_smote).transpose()

# Ajouter l'accuracy au DataFrame une seule fois
accuracy_df_without_smote = pd.DataFrame({'precision': [accuracy_without_smote], 'recall': [accuracy_without_smote], 'f1-score': [accuracy_without_smote], 'support': [len(Y_test)]}, index=['accuracy'])
report_df_without_smote = pd.concat([report_df_without_smote.loc[['0', '1']], accuracy_df_without_smote])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nModèle SVM sans SMOTE")
print(report_df_without_smote)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_without_smote = report_without_smote['0']['f1-score']
f1_score_class_1_without_smote = report_without_smote['1']['f1-score']

# Afficher séparément les F1-scores pour chaque classe
print(f"\nF1-Score (classe 0): {f1_score_class_0_without_smote:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_without_smote:.4f}")

# Calculer le F1-score global (moyenne macro ou micro)
f1_macro_without_smote = f1_score(Y_test, y_pred_without_smote, average='macro')
f1_micro_without_smote = f1_score(Y_test, y_pred_without_smote, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_without_smote:.4f}")
print(f"F1-Score (micro): {f1_micro_without_smote:.4f}")
```

## <span style="color:black;">Avec SMOTE</span>

```python
# Appliquer SMOTE pour augmenter les échantillons de la classe minoritaire
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
# Sélectionner un sous-ensemble de 50 000 lignes aléatoires
X_train_subset_smote, _, Y_train_subset_smote, _ = train_test_split(X_train_smote, Y_train_smote, train_size=50000, random_state=42)
```

```python
# Séparer les échantillons en fonction des classes
fraud_indices = np.where(Y_train_smote == 1)[0]
non_fraud_indices = np.where(Y_train_smote == 0)[0]

# Sélectionner 50 000 lignes aléatoires pour chaque classe
fraud_sample_indices = np.random.choice(fraud_indices, size=50000, replace=False)
non_fraud_sample_indices = np.random.choice(non_fraud_indices, size=50000, replace=False)

# Combiner les indices pour créer l'ensemble final de 100 000 lignes
selected_indices = np.concatenate([fraud_sample_indices, non_fraud_sample_indices])

# Créer les ensembles d'entraînement équilibrés
X_train_subset_balanced = X_train_smote[selected_indices]
Y_train_subset_balanced = Y_train_smote[selected_indices]


X_train_subset_smote = X_train_subset_balanced
Y_train_subset_smote = Y_train_subset_balanced
```

```python
# Initialiser le modèle SVM avec un noyau linéaire
svm_model_with_smote = SVC(kernel='linear', random_state=42)
```

```python
# Entraîner le modèle sur les données d'entraînement (avec SMOTE et limité à 50 000 lignes)
svm_model_with_smote.fit(X_train_subset_smote, Y_train_subset_smote)
```

```python
# Prédire sur les données de test
y_pred_with_smote = svm_model_with_smote.predict(X_test)
```

#### <span style="color:black;">Évaluation des Performances du Modèle d'arbre de décision</span>


```python
# Calculer l'accuracy
accuracy_with_smote = accuracy_score(Y_test, y_pred_with_smote)

# Obtenir le rapport de classification sous forme de dictionnaire
report_with_smote = classification_report(Y_test, y_pred_with_smote, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_with_smote = pd.DataFrame(report_with_smote).transpose()

# Ajouter l'accuracy au DataFrame une seule fois
accuracy_df_with_smote = pd.DataFrame({'precision': [accuracy_with_smote], 'recall': [accuracy_with_smote], 'f1-score': [accuracy_with_smote], 'support': [len(Y_test)]}, index=['accuracy'])
report_df_with_smote = pd.concat([report_df_with_smote.loc[['0', '1']], accuracy_df_with_smote])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nModèle SVM avec SMOTE (50 000 lignes)")
print(report_df_with_smote)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_with_smote = report_with_smote['0']['f1-score']
f1_score_class_1_with_smote = report_with_smote['1']['f1-score']

# Afficher séparément les F1-scores pour chaque classe
print(f"\nF1-Score (classe 0): {f1_score_class_0_with_smote:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_with_smote:.4f}")

# Calculer le F1-score global (moyenne macro ou micro)
f1_macro_with_smote = f1_score(Y_test, y_pred_with_smote, average='macro')
f1_micro_with_smote = f1_score(Y_test, y_pred_with_smote, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_with_smote:.4f}")
print(f"F1-Score (micro): {f1_micro_with_smote:.4f}")
```

## <span style="color:black;">GRAPHES</span>


```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# Pour configurer l'affichage des graphiques avec un style plus moderne
sns.set(style="whitegrid")
```

## <span style="color:black;">Évaluation des Modèles : Courbes ROC avec et sans SMOTE</span>


```python
from sklearn.metrics import roc_curve, roc_auc_score

def plot_roc_curve(ax, y_true, y_pred_prob, title):
    fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
    auc_score = roc_auc_score(y_true, y_pred_prob)

    ax.plot(fpr, tpr, color='blue', label=f'AUC = {auc_score:.2f}')
    ax.plot([0, 1], [0, 1], color='red', linestyle='--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve: {title}')
    ax.legend(loc='lower right')

# Prédictions probabilistes pour les modèles sans et avec SMOTE
y_test_pred_prob_without_smote = svm_model_without_smote.decision_function(X_test)
y_test_pred_prob_with_smote = svm_model_with_smote.decision_function(X_test)

# Créer une figure avec deux sous-graphiques (axes)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 ligne, 2 colonnes

# Tracer la courbe ROC pour le modèle sans SMOTE
plot_roc_curve(axes[0], Y_test, y_test_pred_prob_without_smote, "Modèle SVM sans SMOTE")

# Tracer la courbe ROC pour le modèle avec SMOTE
plot_roc_curve(axes[1], Y_test, y_test_pred_prob_with_smote, "Modèle SVM avec SMOTE")

plt.tight_layout()
plt.show()
```

### Analyse des courbes ROC pour les modèles SVM avec et sans SMOTE

#### Compréhension des courbes ROC:
Les courbes ROC permettent de visualiser le compromis entre la sensibilité (taux de vrais positifs) et la spécificité (1 - taux de faux positifs) pour différents seuils de classification.

#### Observations:
- **AUC plus élevée avec SMOTE**: La courbe ROC du modèle SVM avec SMOTE est au-dessus de celle sans SMOTE, indiquant une meilleure performance du modèle avec SMOTE.


#### Conclusion:
L'utilisation de SMOTE améliore les performances du modèle SVM en termes de détection des fraudes, comme le montre l'augmentation de l'AUC sur la courbe ROC.



## <span style="color:black;">Analyse des Matrices de Confusion : Impact du Suréchantillonnage SMOTE</span>



```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(ax, y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Purples, ax=ax, colorbar=False)  # Désactiver la barre de couleurs pour un look plus épuré
    ax.set_title(f'Matrice de Confusion: {title}')
    ax.grid(False)  # Désactiver la grille
    ax.set_xticks([])  # Supprimer les ticks de l'axe x
    ax.set_yticks([])  # Supprimer les ticks de l'axe y
    ax.spines['top'].set_visible(False)  # Masquer la bordure du haut
    ax.spines['right'].set_visible(False)  # Masquer la bordure de droite
    ax.spines['left'].set_visible(False)  # Masquer la bordure de gauche
    ax.spines['bottom'].set_visible(False)  # Masquer la bordure du bas


# Créer une figure avec deux sous-graphiques (axes)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 ligne, 2 colonnes

# Matrice de confusion pour le modèle sans SMOTE
plot_confusion_matrix(axes[0], Y_test, y_pred_without_smote, "Modèle sans SMOTE")

# Matrice de confusion pour le modèle avec SMOTE
plot_confusion_matrix(axes[1], Y_test, y_pred_with_smote, "Modèle avec SMOTE")

plt.tight_layout()
plt.show()
```

### Analyse des matrices de confusion pour les modèles SVM avec et sans SMOTE

#### Compréhension des matrices de confusion:
Elles permettent de visualiser les erreurs de classification en comparant les prédictions aux valeurs réelles pour chaque catégorie (vrais/faux positifs, vrais/faux négatifs).

#### Observations:
- **Déséquilibre des classes**: La classe négative est largement majoritaire, typique dans les problèmes de fraude.
- **Impact de SMOTE**:
  - **Augmentation des vrais positifs**: Le modèle avec SMOTE détecte mieux la classe minoritaire.
  - **Léger compromis sur les faux positifs**: Le nombre de faux positifs augmente légèrement avec SMOTE.

#### Interprétation:
- **SMOTE améliore le rappel**: Le modèle détecte plus de cas de fraude (classe minoritaire) avec SMOTE, tout en réduisant les faux négatifs.
- **Compromis précision-rappel**: Bien que la précision baisse légèrement avec l'augmentation des faux positifs, le rappel est significativement amélioré.

#### Analyse numérique:
- **Sans SMOTE**:
  - Vrais positifs: 1616, Faux positifs: 0
- **Avec SMOTE**:
  - Vrais positifs: 1200, Faux positifs: 20580

#### Conclusion:
SMOTE améliore les performances pour la détection de la fraude, en augmentant le rappel au détriment d'une légère hausse des faux positifs.

#### Interprétation:

- **MCC proche de 0**: Performances proches du hasard.

L'application de SMOTE a considérablement amélioré les performances du modèle, passant d'un MCC de 0.27 à 0.81. Cela montre une meilleure capacité du modèle à distinguer les transactions frauduleuses des transactions légitimes, tout en réduisant les erreurs de classification.
