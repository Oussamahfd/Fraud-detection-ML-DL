```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA, FastICA
from sklearn.linear_model import LogisticRegression
```

## <span style="color:black;">Chargement des datasets</span>


```python
# Chargement des données
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
```

## <span style="color:black;">Encodage des Variables Catégorielles avec Label Encoding</span>


```python
from sklearn.preprocessing import LabelEncoder

# Créez un dictionnaire pour les encodeurs
label_encoders = {}

# Sélection des colonnes catégorielles
categorical_columns = [col for col in df_train.columns if df_train[col].dtype == 'object']

for col in categorical_columns:
    le = LabelEncoder()
    le.fit(df_train[col].astype(str))  # Ajustez l'encodeur sur le jeu de données d'entraînement
    df_train[col] = le.transform(df_train[col].astype(str))  # Transformez le jeu d'entraînement
    label_encoders[col] = le
```

## <span style="color:black;">Visualisation de la Matrice de Corrélation des Caractéristiques</span>


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sélection des caractéristiques et de la cible
X = df_train.drop(['is_fraud'], axis=1)
y = df_train['is_fraud']

# Calcul de la matrice de corrélation
correlation_matrix = X.corr()

# Affichage de la matrice de corrélation
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de Corrélation')
plt.show()
```

## <span style="color:black;">Entraînement du Modèle de Régression Logistique et Analyse des Coefficients des Caractéristiques</span>


```python
# Entraînement du modèle de régression logistique
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# Affichage des coefficients des caractéristiques
coefficients = pd.Series(model.coef_[0], index=X.columns)
coefficients = coefficients.sort_values(ascending=False)
print("Coefficients des caractéristiques :")
print(coefficients)
```

## <span style="color:black;">Évaluation des Caractéristiques à l'Aide de SelectKBest et ANOVA</span>


```python
from sklearn.feature_selection import SelectKBest, f_classif

# Application de SelectKBest avec ANOVA
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)

# Affichage des scores des caractéristiques
scores = pd.Series(selector.scores_, index=X.columns)
scores = scores.sort_values(ascending=False)
print("Scores des caractéristiques :")
print(scores)
```


## <span style="color:black;">Détection et Gestion des Valeurs Aberrantes</span>



### <span style="color:black;">Détection des Valeurs Aberrantes dans les Colonnes Numériques en Utilisant la Méthode IQR</span>


```python
numeric_columns = df_train.select_dtypes(include=['float64', 'int64']).columns

# Détection des valeurs aberrantes avec IQR
outliers_iqr = {}

for col in numeric_columns:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_train[(df_train[col] < lower_bound) | (df_train[col] > upper_bound)]
    outliers_iqr[col] = outliers

print("Valeurs aberrantes détectées avec IQR :")
for col, outliers in outliers_iqr.items():
    print(f"{col}: {len(outliers)} valeurs aberrantes")
```


### <span style="color:black;">Suppression des Valeurs Aberrantes en Utilisant les Méthodes IQR et Z-Score</span>


```python
# Suppression des valeurs aberrantes en utilisant IQR
df_train_cleaned_iqr = df_train[~df_train.index.isin(pd.concat(outliers_iqr.values()).index)]

# Suppression des valeurs aberrantes en utilisant Z-Score
df_train_cleaned_zscore = df_train[~df_train.index.isin(pd.concat(outliers_zscore.values()).index)]
```

## <span style="color:black;">Analyse de la Variance des Statistiques Descriptives entre Sous-Échantillons</span>



```python
import numpy as np

# Fonction pour calculer les statistiques descriptives
def calculate_statistics(df, features):
    return df[features].describe().T[['mean', 'std']]

# Nombre de sous-échantillons
n_samples = 5

# Créez des sous-échantillons
sub_samples = [df_train.sample(frac=0.8, random_state=i) for i in range(n_samples)]

# Caractéristiques à analyser
features = df_train.columns

# Calcul des statistiques pour chaque sous-échantillon
statistics = [calculate_statistics(sample, features) for sample in sub_samples]

# Calcul de la variance des statistiques entre les sous-échantillons
variances = np.var([stat[['mean']].values for stat in statistics], axis=0)
print("Variance des statistiques des caractéristiques :")
print(variances)
```

### <span style="color:black;">Validation Croisée pour l'Évaluation des Caractéristiques</span>



```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Fonction pour évaluer la performance avec validation croisée
def cross_validate_model(X, y):
    model = LogisticRegression(max_iter=1000)
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    return scores.mean()

# Performance du modèle avec toutes les caractéristiques
accuracy_full = cross_validate_model(X_train, y_train)
print("Précision du modèle avec toutes les caractéristiques :", accuracy_full)
```

### <span style="color:black;">Analyse de la Performance du Modèle avec et sans certaines Caractéristiques</span>



```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Fonction pour évaluer la performance avec validation croisée
def cross_validate_model(X, y):
    model = LogisticRegression(max_iter=1000)  # Vous pouvez choisir un autre modèle
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    return scores.mean()

# Fonction pour évaluer la performance du modèle en supprimant certaines caractéristiques
def evaluate_model_without_features(features_to_remove):
    X_train_reduced = X_train.drop(columns=features_to_remove)
    
    # Performance du modèle réduit avec validation croisée
    accuracy_reduced = cross_validate_model(X_train_reduced, y_train)
    
    return accuracy_reduced

# Liste des caractéristiques à supprimer
features_to_remove = [
    'Unnamed: 0', 'cc_num', 'first', 'last', 'street', 'zip',
    'unix_time', 'dob', 'trans_num', 'lat', 'long', 
    'merch_lat', 'merch_long'
]

# Évaluez la performance sans ces caractéristiques
accuracy_reduced = evaluate_model_without_features(features_to_remove)

# Calculer la précision du modèle complet (avant suppression)
accuracy_full = cross_validate_model(X_train, y_train)

print("Précision du modèle sans certaines caractéristiques :", accuracy_reduced)
print("Différence par rapport au modèle complet :", accuracy_full - accuracy_reduced)

```
