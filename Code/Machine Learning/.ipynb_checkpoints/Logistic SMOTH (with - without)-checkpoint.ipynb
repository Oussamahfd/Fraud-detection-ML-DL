# <span style="color:black;">Importation des Bibliothèques et Préparation des Outils pour l'Analyse
</span>


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report ,f1_score
from imblearn.over_sampling import SMOTE
```

# <span style="color:black;">Préparation et Prétraitement des Données pour la Détection de Fraude
</span>


## <span style="color:black;">Chargement des datasets</span>


```python
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

## <span style="color:black;">Suppression des colonnes inutiles</span>


```python
df_train.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
df_test.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
```

##  <span style="color:black;">Conversion des colonnes de dates en timestamps</span>


```python
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_train['dob'] = pd.to_datetime(df_train['dob']).apply(lambda x: x.timestamp())

df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_test['dob'] = pd.to_datetime(df_test['dob']).apply(lambda x: x.timestamp())
```

##  <span style="color:black;">Encodage des colonnes catégorielles</span>


```python
label_encoder = LabelEncoder()
columns_to_convert_in_string_to_int = ['category', 'city', 'state']
for col in columns_to_convert_in_string_to_int:
    df_train[col] = label_encoder.fit_transform(df_train[col])
    df_test[col] = label_encoder.fit_transform(df_test[col])
```

##  <span style="color:black;">Conversion des colonnes en float</span>


```python
columns_to_convert_in_int_to_float = ['category', 'city', 'state', 'city_pop']
for col in columns_to_convert_in_int_to_float:
    df_train[col] = df_train[col].astype('float64')
    df_test[col] = df_test[col].astype('float64')
```

```python
df_train.info()
```

```python
df_test.info()
```

## <span style="color:black;">Séparation des features et de la target</span>


```python
X = df_train.drop(columns=['is_fraud'])
Y = df_train['is_fraud']
```

##  <span style="color:black;">Normalisation des features</span>


```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

##  <span style="color:black;">Séparation des données en train et test</span>


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

# <span style="color:black;">Entraînement du Modèle de Régression Logistique [Sans / Avec] SMOTE</span>


##  <span style="color:black;">Modèle sans SMOTE</span>


```python
model_without_smote = LogisticRegression(max_iter=1000, random_state=42)
model_without_smote.fit(X_train, Y_train)
y_train_pred_without_smote = model_without_smote.predict(X_train)
```

## <span style="color:black;">Évaluation des Performances du Modèle sans SMOTE</span>


```python
# Calculer l'accuracy
accuracy_without_smote = accuracy_score(Y_train, y_train_pred_without_smote)

# Obtenir le rapport de classification sous forme de dictionnaire
report_without_smote = classification_report(Y_train, y_train_pred_without_smote, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_without_smote = pd.DataFrame(report_without_smote).transpose()

# Ajouter l'accuracy au DataFrame
accuracy_df_without_smote = pd.DataFrame({'precision': [accuracy_without_smote], 
                                          'recall': [accuracy_without_smote], 
                                          'f1-score': [accuracy_without_smote], 
                                          'support': [len(Y_train)]}, 
                                         index=['accuracy'])

# Garder uniquement les classes 0 et 1 et ajouter l'accuracy
report_df_without_smote = pd.concat([report_df_without_smote.loc[['0', '1']], accuracy_df_without_smote])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nModèle sans SMOTE")
print(report_df_without_smote)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_without_smote = report_without_smote['0']['f1-score']
f1_score_class_1_without_smote = report_without_smote['1']['f1-score']

# Afficher les F1-scores
print(f"\nF1-Score (classe 0): {f1_score_class_0_without_smote:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_without_smote:.4f}")

# Calculer les F1-scores globaux (macro et micro)
f1_macro_without_smote = f1_score(Y_train, y_train_pred_without_smote, average='macro')
f1_micro_without_smote = f1_score(Y_train, y_train_pred_without_smote, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_without_smote:.4f}")
print(f"F1-Score (micro): {f1_micro_without_smote:.4f}")

# Afficher l'accuracy
print(f"\nAccuracy: {accuracy_without_smote:.4f}")
```

## <span style="color:black;"> Application de SMOTE pour équilibrer les classes
</span>


```python
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

##  <span style="color:black;">Modèle avec SMOTE</span>


```python
model_with_smote = LogisticRegression(max_iter=1000, random_state=42)
model_with_smote.fit(X_train_smote, Y_train_smote)
y_train_pred_with_smote = model_with_smote.predict(X_test)
```

## <span style="color:black;">Évaluation des Performances du Modèle avec SMOTE</span>


```python
# Calculer l'accuracy
accuracy_with_smote = accuracy_score(Y_train_smote, y_train_pred_with_smote)

# Obtenir le rapport de classification sous forme de dictionnaire
report_with_smote = classification_report(Y_train_smote, y_train_pred_with_smote, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_with_smote = pd.DataFrame(report_with_smote).transpose()

# Ajouter l'accuracy au DataFrame
accuracy_df_with_smote = pd.DataFrame({'precision': [accuracy_with_smote], 
                                       'recall': [accuracy_with_smote], 
                                       'f1-score': [accuracy_with_smote], 
                                       'support': [len(Y_train_smote)]}, 
                                      index=['accuracy'])

# Garder uniquement les classes 0 et 1 et ajouter l'accuracy
report_df_with_smote = pd.concat([report_df_with_smote.loc[['0', '1']], accuracy_df_with_smote])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nModèle avec SMOTE")
print(report_df_with_smote)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_with_smote = report_with_smote['0']['f1-score']
f1_score_class_1_with_smote = report_with_smote['1']['f1-score']

# Afficher les F1-scores
print(f"\nF1-Score (classe 0): {f1_score_class_0_with_smote:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_with_smote:.4f}")

# Calculer les F1-scores globaux (macro et micro)
f1_macro_with_smote = f1_score(Y_train_smote, y_train_pred_with_smote, average='macro')
f1_micro_with_smote = f1_score(Y_train_smote, y_train_pred_with_smote, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_with_smote:.4f}")
print(f"F1-Score (micro): {f1_micro_with_smote:.4f}")

# Afficher l'accuracy
print(f"\nAccuracy: {accuracy_with_smote:.4f}")
```

# <span style="color:black;">Evaluation du Modèle de Régression Logistique [Sans / Avec] SMOTE</span>


## <span style="color:black;">GRAPHES</span>


```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# Pour configurer l'affichage des graphiques avec un style plus moderne
sns.set(style="whitegrid")
```

## <span style="color:black;">Évaluation des Modèles : Courbes ROC avec et sans SMOTE</span>


```python
from sklearn.metrics import roc_curve, roc_auc_score

def plot_roc_curve(ax, y_true, y_pred_prob, title):
    fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
    auc_score = roc_auc_score(y_true, y_pred_prob)

    ax.plot(fpr, tpr, color='blue', label=f'AUC = {auc_score:.2f}')
    ax.plot([0, 1], [0, 1], color='red', linestyle='--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve: {title}')
    ax.legend(loc='lower right')

# Prédictions probabilistes pour les modèles sans et avec SMOTE
y_train_pred_prob_without_smote = model_without_smote.predict_proba(X_train)[:, 1]
y_train_pred_prob_with_smote = model_with_smote.predict_proba(X_train_smote)[:, 1]

# Créer une figure avec deux sous-graphiques (axes)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 ligne, 2 colonnes

# Tracer la courbe ROC pour le modèle sans SMOTE
plot_roc_curve(axes[0], Y_train, y_train_pred_prob_without_smote, "Modèle sans SMOTE")

# Tracer la courbe ROC pour le modèle avec SMOTE
plot_roc_curve(axes[1], Y_train_smote, y_train_pred_prob_with_smote, "Modèle avec SMOTE")

plt.tight_layout()
plt.show()
```

### Interprétation des résultats

#### Comparaison des deux modèles
Les deux graphiques montrent que les deux modèles ont des performances globalement bonnes, car leurs courbes sont bien au-dessus de la diagonale. Cependant, le modèle avec SMOTE présente une légère amélioration par rapport au modèle sans SMOTE, comme en témoigne l'AUC légèrement supérieure (0.86 contre 0.85).

#### Impact de SMOTE
L'utilisation de SMOTE, une technique d'oversampling, semble avoir permis d'améliorer légèrement les performances du modèle en réduisant le déséquilibre entre les classes (si tel était le cas dans vos données). Cela suggère que SMOTE a aidé le modèle à mieux généraliser et à réduire le biais lié à la classe minoritaire.

### Conclusion
En résumé, les graphiques ROC montrent que les deux modèles sont performants, mais que le modèle utilisant SMOTE offre de légèrement meilleurs résultats. Cela suggère que l'utilisation de SMOTE peut être bénéfique dans les cas où les données présentent un déséquilibre de classe.


## <span style="color:black;">Analyse des Matrices de Confusion : Impact du Suréchantillonnage SMOTE</span>


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(ax, y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Purples, ax=ax, colorbar=False)  # Désactiver la barre de couleurs pour un look plus épuré
    ax.set_title(f'Matrice de Confusion: {title}')
    ax.grid(False)  # Désactiver la grille
    ax.set_xticks([])  # Supprimer les ticks de l'axe x
    ax.set_yticks([])  # Supprimer les ticks de l'axe y
    ax.spines['top'].set_visible(False)  # Masquer la bordure du haut
    ax.spines['right'].set_visible(False)  # Masquer la bordure de droite
    ax.spines['left'].set_visible(False)  # Masquer la bordure de gauche
    ax.spines['bottom'].set_visible(False)  # Masquer la bordure du bas

# Prédictions finales pour les modèles sans et avec SMOTE
y_train_pred_final_without_smote = model_without_smote.predict(X_train)
y_train_pred_final_with_smote = model_with_smote.predict(X_train_smote)

# Créer une figure avec deux sous-graphiques (axes)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 ligne, 2 colonnes

# Matrice de confusion pour le modèle sans SMOTE
plot_confusion_matrix(axes[0], Y_train, y_train_pred_final_without_smote, "Modèle sans SMOTE")

# Matrice de confusion pour le modèle avec SMOTE
plot_confusion_matrix(axes[1], Y_train_smote, y_train_pred_final_with_smote, "Modèle avec SMOTE")

plt.tight_layout()
plt.show()
```

### Interprétation des résultats

- **Déséquilibre de classe accentué** : Les matrices de confusion montrent que le déséquilibre entre les classes est encore plus marqué. La classe majoritaire est correctement prédite en grande majorité, ce qui se traduit par des valeurs élevées dans la première ligne et colonne.
  
- **Impact de SMOTE** :
  - **Amélioration pour la classe minoritaire** : Le modèle avec SMOTE prédit mieux la classe minoritaire, comme le montre une valeur plus élevée dans la case en bas à droite de la matrice.
  - **Augmentation des faux positifs** : Cependant, on observe une augmentation des faux positifs (en haut à droite de la matrice avec SMOTE), ce qui signifie que le modèle classe davantage d'exemples de la classe majoritaire comme minoritaires.

- **Compromis entre précision et rappel** : L'amélioration du rappel pour la classe minoritaire se fait au prix d'une précision légèrement inférieure, illustrant un compromis entre ces deux métriques.

### Conclusion

SMOTE s'avère efficace pour améliorer la détection de la classe minoritaire en réduisant le biais dû au déséquilibre des classes. Cependant, ce gain se fait au détriment d'une augmentation des faux positifs, ce qui nécessite de bien évaluer ce compromis. Il est recommandé de poursuivre l'analyse avec des métriques supplémentaires telles que le F1-score, les courbes ROC et de précision-rappel pour mieux comprendre l'impact global de SMOTE sur les performances du modèle.


## <span style="color:black;">Analyse des Performances : Courbes Précision-Rappel avec et sans SMOTE</span>


```python
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_precision_recall_curve(ax, y_true, y_pred_prob, title):
    precision, recall, _ = precision_recall_curve(y_true, y_pred_prob)
    avg_precision = average_precision_score(y_true, y_pred_prob)

    ax.plot(recall, precision, color='blue', label=f'Avg Precision = {avg_precision:.2f}')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    ax.set_title(f'Precision-Recall Curve: {title}')
    ax.legend(loc='lower left')

# Prédictions probabilistes pour les modèles sans et avec SMOTE
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 ligne, 2 colonnes

plot_precision_recall_curve(axes[0], Y_train, y_train_pred_prob_without_smote, "Modèle sans SMOTE")
plot_precision_recall_curve(axes[1], Y_train_smote, y_train_pred_prob_with_smote, "Modèle avec SMOTE")

plt.tight_layout()
plt.show()
```

### Interprétation des résultats

- **Modèle sans SMOTE** : La courbe de précision-rappel présente une forme en cloche, avec une aire sous la courbe relativement faible. Cela indique que le modèle a du mal à maintenir une précision élevée tout en atteignant un rappel élevé, en particulier pour la classe minoritaire.

- **Modèle avec SMOTE** : La courbe de précision-rappel est nettement meilleure, avec une aire sous la courbe beaucoup plus grande. Le modèle parvient à maintenir à la fois une précision et un rappel élevés, notamment pour la classe minoritaire, grâce à l'utilisation de SMOTE.

### Conclusion

L'application de SMOTE a permis d'améliorer significativement les performances de la classification pour la classe minoritaire, en réduisant le biais dû au déséquilibre des classes. Le choix du seuil de classification peut être optimisé en fonction des besoins spécifiques de l'application (privilégier un rappel élevé ou une précision élevée). Pour aller plus loin, des ajustements d'hyperparamètres, d'autres techniques de gestion de l'équilibre des classes, et une meilleure ingénierie des caractéristiques peuvent être explorés pour optimiser davantage les résultats.

## <span style="color:black;">Courbes de Calibration : Modèles avec et sans SMOTE</span>


```python
from sklearn.calibration import calibration_curve

def plot_calibration_curve(ax, y_true, y_pred_prob, title):
    prob_true, prob_pred = calibration_curve(y_true, y_pred_prob, n_bins=10)
    ax.plot(prob_pred, prob_true, marker='o', label='Calibration Curve')
    ax.plot([0, 1], [0, 1], linestyle='--', color='gray')
    ax.set_title(f'Calibration Curve: {title}')
    ax.set_xlabel('Predicted Probability')
    ax.set_ylabel('True Probability')
    ax.legend()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

plot_calibration_curve(axes[0], Y_train, y_train_pred_prob_without_smote, "Modèle sans SMOTE")
plot_calibration_curve(axes[1], Y_train_smote, y_train_pred_prob_with_smote, "Modèle avec SMOTE")

plt.tight_layout()
plt.show()
```

### Interprétation des résultats

- **Modèle sans SMOTE** :
  - La courbe d'étalonnage s'écarte de la diagonale à 45 degrés, indiquant un mauvais calibrage.
  - Surestimation des probabilités réelles pour les faibles valeurs prédites et sous-estimation pour les valeurs élevées.

- **Modèle avec SMOTE** :
  - La courbe est plus proche de la diagonale, signalant un meilleur calibrage.
  - Les probabilités prédites reflètent mieux les probabilités réelles des événements.

### Conclusion

L'utilisation de SMOTE a amélioré l'étalonnage du modèle, rendant ses prédictions de probabilité plus fiables, crucial pour des applications nécessitant la certitude des prédictions.
