```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# CrÃ©er une copie de df_train pour les transformations
df_corr = df_train.copy()

# Instantiate the LabelEncoder
encoder = LabelEncoder()

# Appliquer les transformations sur df_corr sans modifier df_train
df_corr["category"] = encoder.fit_transform(df_corr["category"])
df_corr["gender"] = encoder.fit_transform(df_corr["gender"])
df_corr["city"] = encoder.fit_transform(df_corr["city"])
df_corr["state"] = encoder.fit_transform(df_corr["state"])
df_corr["job"] = encoder.fit_transform(df_corr["job"])
```

```python
# Select only numeric columns
numeric_df = df_corr.select_dtypes(include='number')

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix, cmap="Reds", annot=True, fmt=".2f")
plt.show()
```

```python
df_train .drop(columns=['Unnamed: 0','merchant', 'cc_num','first', 'last','gender','trans_num','unix_time','street','merch_lat','merch_long','job','zip'], inplace= True)
```

```python
df_train.head()
```

```python
# convert all dtype in float
df_train ['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
df_train ['trans_date_trans_time'] = df_train ['trans_date_trans_time'].apply(lambda x: x.timestamp())
df_train ['dob'] = pd.to_datetime(df_train ['dob'])
df_train ['dob'] = df_train ['dob'].apply(lambda x: x.timestamp())
label_encoder = LabelEncoder()
columns_to_convert_in_string_to_int = ['category','city', 'state']
for col in columns_to_convert_in_string_to_int:
    df_train [col] = label_encoder.fit_transform(df_train [col])
columns_to_convert_in_int_to_float = ['category','city', 'state','city_pop']
for col in columns_to_convert_in_int_to_float:
    df_train [col] = df_train [col].astype('float64')
```

```python
X = df_train .drop(columns=['is_fraud'])
Y = df_train ['is_fraud']
```

```python
scaler = StandardScaler()
X= scaler.fit_transform(X)
```

```python
X_train, _ , Y_train, _ = train_test_split(X, Y, random_state=42)
```

```python
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

```python
df_test .drop(columns=['Unnamed: 0','merchant', 'cc_num','first', 'last','gender','trans_num','unix_time','street','merch_lat','merch_long','job','zip'], inplace= True)
```

```python
# convert all dtype in float
df_test ['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test ['trans_date_trans_time'] = df_test ['trans_date_trans_time'].apply(lambda x: x.timestamp())
df_test ['dob'] = pd.to_datetime(df_test ['dob'])
df_test ['dob'] = df_test ['dob'].apply(lambda x: x.timestamp())
```

```python
label_encoder = LabelEncoder()
columns_to_convert = ['category','city', 'state']
for col in columns_to_convert:
    df_test [col] = label_encoder.fit_transform(df_test [col])
```

```python
columns_to_convert = ['category','city', 'state','city_pop']
for col in columns_to_convert:
    df_test [col] = df_test [col].astype('float64')
```

```python
df_test.dtypes
```

```python
X_test_initial = df_test .drop(columns=['is_fraud'])
Y_test_initial = df_test ['is_fraud']
```

```python
scaler = StandardScaler()
X_test_initial= scaler.fit_transform(X_test_initial)
```

```python
X_test, _ , Y_test, _ = train_test_split(X_test_initial, Y_test_initial, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

# Appliquer SMOTE pour Ã©quilibrer les classes sur l'ensemble d'entraÃ®nement
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)

# Normaliser les donnÃ©es (normalisation standard)
scaler = StandardScaler()
X_train_smote = scaler.fit_transform(X_train_smote)
X_test = scaler.transform(X_test)  # Utiliser le mÃªme scaler pour les donnÃ©es de test


```

```python
# Initialiser et entraÃ®ner le modÃ¨le KNN
knn_model = KNeighborsClassifier(n_neighbors=5)  # Vous pouvez ajuster le nombre de voisins K ici
knn_model.fit(X_train_smote, Y_train_smote)

# Faire des prÃ©dictions sur les donnÃ©es de test
knn_predictions = knn_model.predict(X_test)

# Calculer l'exactitude du modÃ¨le
knn_accuracy = accuracy_score(Y_test, knn_predictions)

# Afficher les mÃ©triques d'Ã©valuation
print("KNN Accuracy: {:.3f}%".format(knn_accuracy * 100))
print("\nClassification Report:\n", classification_report(Y_test, knn_predictions))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, knn_predictions))
```

```python
# Initialiser et entraÃ®ner le modÃ¨le KNN
knn_model = KNeighborsClassifier(n_neighbors=3)  # Vous pouvez ajuster le nombre de voisins K ici
knn_model.fit(X_train_smote, Y_train_smote)

# Faire des prÃ©dictions sur les donnÃ©es de test
knn_predictions = knn_model.predict(X_test)

# Calculer l'exactitude du modÃ¨le
knn_accuracy = accuracy_score(Y_test, knn_predictions)

# Afficher les mÃ©triques d'Ã©valuation
print("KNN Accuracy: {:.3f}%".format(knn_accuracy * 100))
print("\nClassification Report:\n", classification_report(Y_test, knn_predictions))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, knn_predictions))
```
