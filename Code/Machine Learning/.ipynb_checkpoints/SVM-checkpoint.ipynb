<h1>Jusification a supprimer gneder</h1>

Faible Variance de la Variable : Si la différence entre les taux de fraude par genre est négligeable, la variable gender peut avoir peu ou pas de valeur prédictive. Dans ce cas, vous pourriez envisager de supprimer cette variable, car elle n'apporte pas d'information supplémentaire significative à votre modèle.


Évitement de la Redondance : Si d'autres variables capturent déjà l'essentiel des informations liées à la fraude (par exemple, le montant de la transaction, la localisation, etc.), gender pourrait être redondant.


Prévenir le Biais : Si l'inclusion de la variable gender pourrait introduire un biais non souhaité (par exemple, si le modèle surévalue ou sous-évalue la probabilité de fraude en fonction du genre), vous pourriez également choisir de la supprimer pour des raisons éthiques.

```python
# Compter le nombre de fraudes et de non-fraudes pour chaque genre
fraud_count_by_gender = df_train.groupby(['gender', 'is_fraud']).size().unstack()

# Afficher le résultat
print(fraud_count_by_gender)
```

```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Sélectionner uniquement les colonnes numériques
numeric_columns = df_train.select_dtypes(include=['int64', 'float64']).columns

# Créer un nouveau DataFrame avec uniquement les colonnes numériques
X_numeric = df_train[numeric_columns]

# Initialiser VarianceThreshold avec un seuil de variance
selector = VarianceThreshold(threshold=0.1)

# Appliquer la transformation pour supprimer les caractéristiques à faible variance
X_reduced = selector.fit_transform(X_numeric)

# Créer un DataFrame avec les colonnes retenues
X_reduced_df = pd.DataFrame(X_reduced, columns=numeric_columns[selector.get_support()])

# Afficher les colonnes conservées
print("Colonnes retenues après VarianceThreshold :\n", X_reduced_df.columns)
```

```python
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

```python
df_test .drop(columns=['Unnamed: 0', 'cc_num', 'dob', 'trans_num','first', 'last','street','zip','unix_time','job','gender'], inplace= True)
df_test
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = ['merchant', 'category', 'city', 'state']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_test[col] = encoder.fit_transform(df_test[col])

# The rest of your numeric columns should already be in the correct format

df_test ['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test ['trans_date_trans_time'] = df_test ['trans_date_trans_time'].apply(lambda x: x.timestamp())
```

```python
df_test.dtypes
```

```python
X_test_initial = df_test .drop(columns=['is_fraud'])
Y_test_initial = df_test ['is_fraud']
```

```python
scaler = StandardScaler()
X_test_initial= scaler.fit_transform(X_test_initial)
```

```python
X_test, _ , Y_test, _ = train_test_split(X_test_initial, Y_test_initial, random_state=42)

```

```python
import time
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialiser le modèle SVM avec un noyau linéaire pour une exécution plus rapide
svm_model = SVC(kernel='linear', random_state=42)

# Mesurer le temps d'entraînement
start_time = time.time()

# Entraîner le modèle sur les données d'entraînement
svm_model.fit(X_train_subset, Y_train_subset)

# Calculer le temps d'entraînement
training_time = time.time() - start_time

# Afficher le temps d'entraînement
print(f"Temps d'entraînement du modèle SVM : {training_time:.4f} secondes")
```

```python
svm_predictions = svm_model.predict(X_test)

# Calculer l'accuracy
svm_accuracy = accuracy_score(Y_test, svm_predictions)

# Afficher l'accuracy
print("SVM Accuracy on Test Data: {:.3f}%".format(svm_accuracy * 100))
```
