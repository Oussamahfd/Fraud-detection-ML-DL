```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
```

```python
df_train .drop(columns=['Unnamed: 0', 'cc_num', 'dob', 'trans_num','first', 'last','street','zip','unix_time','job','gender'], inplace= True)
df_train
```

```python
unique_values_count = df_train['trans_date_trans_time'].nunique()
print(unique_values_count)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = ['merchant', 'category', 'city', 'state']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_train[col] = encoder.fit_transform(df_train[col])

# The rest of your numeric columns should already be in the correct format
```

```python
df_train ['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
df_train ['trans_date_trans_time'] = df_train ['trans_date_trans_time'].apply(lambda x: x.timestamp())
```

```python
unique_values_count = df_train['trans_date_trans_time'].nunique()
print(unique_values_count)
```

```python
X = df_train .drop(columns=['is_fraud'])
Y = df_train ['is_fraud']
```

```python
X_train, _ , Y_train, _ = train_test_split(X, Y, random_state=42)
```

```python
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

```python
df_test .drop(columns=['Unnamed: 0', 'cc_num', 'dob', 'trans_num','first', 'last','street','zip','unix_time','job','gender'], inplace= True)
df_test
```

```python
unique_values_count = df_test['trans_date_trans_time'].nunique()
print(unique_values_count)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = ['merchant', 'category', 'city', 'state']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_test[col] = encoder.fit_transform(df_test[col])

# The rest of your numeric columns should already be in the correct format

df_test ['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test ['trans_date_trans_time'] = df_test ['trans_date_trans_time'].apply(lambda x: x.timestamp())
```

```python
unique_values_count = df_test['trans_date_trans_time'].nunique()
print(unique_values_count)
```

```python
X_test_initial = df_test .drop(columns=['is_fraud'])
Y_test_initial = df_test ['is_fraud']
```

```python
X_test, _ , Y_test, _ = train_test_split(X_test_initial, Y_test_initial, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, Y_train)
```

```python

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix# Predict
y_pred_forest = random_forest.predict(X_test)
print(f"Accuracy: {accuracy_score(Y_test, y_pred_forest)}")
print(classification_report(Y_test, y_pred_forest))
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Entraîner un modèle Random Forest
model_rf = RandomForestClassifier()
model_rf.fit(X_train, Y_train)

# Extraire l'importance des caractéristiques
importances = model_rf.feature_importances_

# Afficher l'importance des caractéristiques
features = X_train.columns
indices = np.argsort(importances)

plt.figure(figsize=(10, 6))
plt.title("Importance des caractéristiques")
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Importance relative')
plt.show()
```
