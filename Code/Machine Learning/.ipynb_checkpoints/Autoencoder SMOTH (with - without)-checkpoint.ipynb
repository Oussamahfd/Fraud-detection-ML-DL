# <span style="color:black;">Importation des Bibliothèques et Préparation des Outils pour l'Analyse
</span>


```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
```

# <span style="color:black;">Préparation et Prétraitement des Données pour la Détection de Fraude
</span>


## <span style="color:black;">Chargement des datasets</span>


```python
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

## <span style="color:black;">Suppression des colonnes inutiles</span>


```python
df_train.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
df_test.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
```

## <span style="color:black;">Conversion et Encodage des Colonnes</span>


```python
# Conversion des colonnes de dates en timestamps
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_train['dob'] = pd.to_datetime(df_train['dob']).apply(lambda x: x.timestamp())
```

```python
# Initialize the LabelEncoder
encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = [ 'category', 'city', 'state','dob']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_train[col] = encoder.fit_transform(df_train[col])
```

```python
# List of columns to encode
columns_to_encode = ['category', 'city', 'state','dob']

# Apply LabelEncoder to each column in the list
for col in columns_to_encode:
    df_test[col] = encoder.fit_transform(df_test[col])
```

```python
# The rest of your numeric columns should already be in the correct format
df_test ['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test ['trans_date_trans_time'] = df_test ['trans_date_trans_time'].apply(lambda x: x.timestamp())
```


## <span style="color:black;">Séparation des features et de la target</span>


```python
X = df_train .drop(columns=['is_fraud'])
Y = df_train ['is_fraud']
```


## <span style="color:black;">Normalisation des features</span>


```python
scaler = StandardScaler()
X= scaler.fit_transform(X)
```


## <span style="color:black;">Séparation des données en train et test</span>


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

## <span style="color:black;"> Division du jeu de données de test normalisé en ensembles de test et de validation


</span>


```python
# Définir l'architecture de l'Autoencoder
input_dim = X_train.shape[1]
encoding_dim = 14  # Vous pouvez ajuster cette dimension

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="tanh")(input_layer)
encoder = Dense(int(encoding_dim / 2), activation="tanh")(encoder)

decoder = Dense(int(encoding_dim / 2), activation='tanh')(encoder)
decoder = Dense(input_dim, activation='sigmoid')(decoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
```

# <span style="color:black;">Entraînement du Modèle de SVM [Sans / Avec] SMOTE</span>


## <span style="color:black;">Sans SMOTE</span>

```python

# Compiler l'Autoencoder
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

```

```python
# Entraîner l'Autoencoder
autoencoder.fit(X_train, X_train, epochs=10, batch_size=32, shuffle=True, validation_split=0.2)

```

```python
# Obtenir les erreurs de reconstruction
X_test_pred = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - X_test_pred, 2), axis=1)
```

```python
# Définir un seuil pour la détection de fraude (vous pouvez ajuster ce seuil)
threshold = np.percentile(mse, 95)  # Par exemple, 95e percentile
```

```python
# Prédictions basées sur le seuil
y_pred_autoencoder = (mse > threshold).astype(int)
```

#### <span style="color:black;">Évaluation des Performances du Modèle d' Autoencoder</span>


```python
# Calculer l'accuracy
accuracy_without_smote_autoencoder = accuracy_score(Y_test, y_pred_autoencoder)

# Obtenir le rapport de classification sous forme de dictionnaire
report_without_smote_autoencoder = classification_report(Y_test, y_pred_autoencoder, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_without_smote_autoencoder = pd.DataFrame(report_without_smote_autoencoder).transpose()

# Ajouter l'accuracy au DataFrame
accuracy_df_without_smote_autoencoder = pd.DataFrame({'precision': [accuracy_without_smote_autoencoder], 
                                                      'recall': [accuracy_without_smote_autoencoder], 
                                                      'f1-score': [accuracy_without_smote_autoencoder], 
                                                      'support': [len(Y_test)]}, 
                                                     index=['accuracy'])
# Garder uniquement les classes 0 et 1 et ajouter l'accuracy
report_df_without_smote_autoencoder = pd.concat([report_df_without_smote_autoencoder.loc[['0', '1']], accuracy_df_without_smote_autoencoder])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nAutoencoder sans SMOTE")
print(report_df_without_smote_autoencoder)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_without_smote_autoencoder = report_without_smote_autoencoder['0']['f1-score']
f1_score_class_1_without_smote_autoencoder = report_without_smote_autoencoder['1']['f1-score']

# Afficher les F1-scores
print(f"\nF1-Score (classe 0): {f1_score_class_0_without_smote_autoencoder:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_without_smote_autoencoder:.4f}")

# Calculer les F1-scores globaux (macro et micro)
f1_macro_without_smote_autoencoder = f1_score(Y_test, y_pred_autoencoder, average='macro')
f1_micro_without_smote_autoencoder = f1_score(Y_test, y_pred_autoencoder, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_without_smote_autoencoder:.4f}")
print(f"F1-Score (micro): {f1_micro_without_smote_autoencoder:.4f}")
```

## <span style="color:black;">Avec SMOTE</span>

```python
# Appliquer SMOTE pour équilibrer les classes
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
# Entraîner un Autoencoder sur les données équilibrées avec SMOTE
autoencoder_smote = Model(inputs=input_layer, outputs=decoder)
```

```python
# Compiler l'Autoencoder
autoencoder_smote.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Entraîner l'Autoencoder avec SMOTE
autoencoder_smote.fit(X_train_smote, X_train_smote, epochs=10, batch_size=32, shuffle=True, validation_split=0.2)
```

```python
# Obtenir les erreurs de reconstruction pour les données de test
X_test_pred_smote = autoencoder_smote.predict(X_test)
mse_smote = np.mean(np.power(X_test - X_test_pred_smote, 2), axis=1)
```

```python
# Utiliser le même seuil pour la détection de fraude
y_pred_autoencoder_smote = (mse_smote > threshold).astype(int)
```

#### <span style="color:black;">Évaluation des Performances du Modèle d' Autoencoder</span>


```python
# Calculer l'accuracy
accuracy_with_smote_autoencoder = accuracy_score(Y_test, y_pred_autoencoder_smote)

# Obtenir le rapport de classification sous forme de dictionnaire
report_with_smote_autoencoder = classification_report(Y_test, y_pred_autoencoder_smote, output_dict=True)

# Convertir le rapport de classification en DataFrame
report_df_with_smote_autoencoder = pd.DataFrame(report_with_smote_autoencoder).transpose()

# Ajouter l'accuracy au DataFrame
accuracy_df_with_smote_autoencoder = pd.DataFrame({'precision': [accuracy_with_smote_autoencoder], 
                                                   'recall': [accuracy_with_smote_autoencoder], 
                                                   'f1-score': [accuracy_with_smote_autoencoder], 
                                                   'support': [len(Y_test)]}, 
                                                  index=['accuracy'])
# Garder uniquement les classes 0 et 1 et ajouter l'accuracy
report_df_with_smote_autoencoder = pd.concat([report_df_with_smote_autoencoder.loc[['0', '1']], accuracy_df_with_smote_autoencoder])

# Afficher le tableau avec uniquement les 3 lignes (classe 0, classe 1, accuracy)
print("\nAutoencoder avec SMOTE")
print(report_df_with_smote_autoencoder)

# Extraire les F1-scores pour les classes 0 et 1
f1_score_class_0_with_smote_autoencoder = report_with_smote_autoencoder['0']['f1-score']
f1_score_class_1_with_smote_autoencoder = report_with_smote_autoencoder['1']['f1-score']

# Afficher les F1-scores
print(f"\nF1-Score (classe 0): {f1_score_class_0_with_smote_autoencoder:.4f}")
print(f"F1-Score (classe 1): {f1_score_class_1_with_smote_autoencoder:.4f}")

# Calculer les F1-scores globaux (macro et micro)
f1_macro_with_smote_autoencoder = f1_score(Y_test, y_pred_autoencoder_smote, average='macro')
f1_micro_with_smote_autoencoder = f1_score(Y_test, y_pred_autoencoder_smote, average='micro')

# Afficher les F1-scores globaux
print(f"\nF1-Score (macro): {f1_macro_with_smote_autoencoder:.4f}")
print(f"F1-Score (micro): {f1_micro_with_smote_autoencoder:.4f}")
```

# <span style="color:black;">Evaluation du Modèle de Régression Logistique [Sans / Avec] SMOTE</span>


## <span style="color:black;">Évaluation des Modèles : Courbes ROC avec et sans SMOTE</span>


```python
from sklearn.metrics import roc_curve, auc

# Courbe ROC pour le modèle sans SMOTE
fpr_without_smote, tpr_without_smote, _ = roc_curve(Y_test, y_pred_autoencoder)
roc_auc_without_smote = auc(fpr_without_smote, tpr_without_smote)

# Courbe ROC pour le modèle avec SMOTE
fpr_with_smote, tpr_with_smote, _ = roc_curve(Y_test, y_pred_autoencoder_smote)
roc_auc_with_smote = auc(fpr_with_smote, tpr_with_smote)

# Créer la figure pour afficher les deux courbes côte à côte
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Tracer la courbe ROC pour le modèle sans SMOTE
axes[0].plot(fpr_without_smote, tpr_without_smote, color='blue', lw=2, label=f'ROC AUC = {roc_auc_without_smote:.2f}')
axes[0].plot([0, 1], [0, 1], color='gray', linestyle='--')
axes[0].set_title('Courbe ROC - Modèle sans SMOTE')
axes[0].set_xlabel('Taux de Faux Positifs (FPR)')
axes[0].set_ylabel('Taux de Vrais Positifs (TPR)')
axes[0].legend(loc='lower right')

# Tracer la courbe ROC pour le modèle avec SMOTE
axes[1].plot(fpr_with_smote, tpr_with_smote, color='green', lw=2, label=f'ROC AUC = {roc_auc_with_smote:.2f}')
axes[1].plot([0, 1], [0, 1], color='gray', linestyle='--')
axes[1].set_title('Courbe ROC - Modèle avec SMOTE')
axes[1].set_xlabel('Taux de Faux Positifs (FPR)')
axes[1].set_ylabel('Taux de Vrais Positifs (TPR)')
axes[1].legend(loc='lower right')

# Afficher les graphes
plt.tight_layout()
plt.show()
```

## <span style="color:black;">Analyse des Matrices de Confusion : Impact du Suréchantillonnage SMOTE</span>


```python
from sklearn.metrics import confusion_matrix

# Calcul des matrices de confusion
cm_without_smote = confusion_matrix(Y_test, y_pred_autoencoder)
cm_with_smote = confusion_matrix(Y_test, y_pred_autoencoder_smote)

# Créer la figure pour afficher les deux matrices côte à côte
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Matrice de confusion pour le modèle sans SMOTE
sns.heatmap(cm_without_smote, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title('Matrice de Confusion - Modèle sans SMOTE')
axes[0].set_xlabel('Prédiction')
axes[0].set_ylabel('Vérité terrain')

# Matrice de confusion pour le modèle avec SMOTE
sns.heatmap(cm_with_smote, annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title('Matrice de Confusion - Modèle avec SMOTE')
axes[1].set_xlabel('Prédiction')
axes[1].set_ylabel('Vérité terrain')

# Afficher les graphes
plt.tight_layout()
plt.show()
```

## <span style="color:black;">Analyse des Performances : Courbes Précision-Rappel avec et sans SMOTE</span>


```python
from sklearn.metrics import precision_recall_curve

# Courbe Précision-Rappel pour le modèle sans SMOTE
precision_without_smote, recall_without_smote, _ = precision_recall_curve(Y_test, mse)
# Courbe Précision-Rappel pour le modèle avec SMOTE
precision_with_smote, recall_with_smote, _ = precision_recall_curve(Y_test, mse_smote)

# Créer la figure pour afficher les deux courbes côte à côte
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Courbe Précision-Rappel pour le modèle sans SMOTE
axes[0].plot(recall_without_smote, precision_without_smote, color='b', label='Précision-Rappel (sans SMOTE)')
axes[0].set_title('Précision-Rappel - Modèle sans SMOTE')
axes[0].set_xlabel('Rappel')
axes[0].set_ylabel('Précision')
axes[0].legend(loc="best")

# Courbe Précision-Rappel pour le modèle avec SMOTE
axes[1].plot(recall_with_smote, precision_with_smote, color='g', label='Précision-Rappel (avec SMOTE)')
axes[1].set_title('Précision-Rappel - Modèle avec SMOTE')
axes[1].set_xlabel('Rappel')
axes[1].set_ylabel('Précision')
axes[1].legend(loc="best")

# Afficher les graphes
plt.tight_layout()
plt.show()
```

## <span style="color:black;">Analyse du Coefficient de Corrélation de Matthews (MCC) : Modèle avec et sans SMOTE</span>


```python
from sklearn.metrics import matthews_corrcoef

# Calculer le MCC pour le modèle sans SMOTE
mcc_without_smote = matthews_corrcoef(Y_test, y_pred_autoencoder)

# Calculer le MCC pour le modèle avec SMOTE
mcc_with_smote = matthews_corrcoef(Y_test, y_pred_autoencoder_smote)

# Préparer les données pour le graphique
mcc_scores = [mcc_without_smote, mcc_with_smote]
labels = ['Sans SMOTE', 'Avec SMOTE']

# Tracer les résultats du MCC
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(labels, mcc_scores, color=['blue', 'green'])
ax.set_title('Coefficient de Corrélation de Matthews (MCC)')
ax.set_xlabel('Modèle')
ax.set_ylabel('MCC')
ax.set_ylim([0, 1])  # Limiter l'axe des Y entre 0 et 1 pour une meilleure comparaison
plt.show()
```

## <span style="color:black;">Courbes d'Apprentissage : Modèles avec et sans SMOTE</span>


```python
import matplotlib.pyplot as plt

# Entraînement de l'Autoencoder sans SMOTE
history_without_smote = autoencoder.fit(X_train, X_train, 
                                        epochs=10, 
                                        batch_size=32, 
                                        shuffle=True, 
                                        validation_split=0.2, 
                                        verbose=0)

# Entraînement de l'Autoencoder avec SMOTE
history_with_smote = autoencoder_smote.fit(X_train_smote, X_train_smote, 
                                           epochs=10, 
                                           batch_size=32, 
                                           shuffle=True, 
                                           validation_split=0.2, 
                                           verbose=0)

# Créer la figure pour afficher les courbes d'apprentissage côte à côte
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Courbes d'apprentissage sans SMOTE
axes[0].plot(history_without_smote.history['loss'], label='Loss - Train')
axes[0].plot(history_without_smote.history['val_loss'], label='Loss - Validation')
axes[0].set_title("Courbes d'Apprentissage sans SMOTE")
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Loss')
axes[0].legend()

# Courbes d'apprentissage avec SMOTE
axes[1].plot(history_with_smote.history['loss'], label='Loss - Train')
axes[1].plot(history_with_smote.history['val_loss'], label='Loss - Validation')
axes[1].set_title("Courbes d'Apprentissage avec SMOTE")
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Loss')
axes[1].legend()

# Afficher les graphes
plt.tight_layout()
plt.show()
```

## <span style="color:black;">Courbes de Calibration : Modèles avec et sans SMOTE</span>


```python
from sklearn.calibration import calibration_curve

# Calculer les courbes de calibration pour le modèle sans SMOTE
mse_normalized = (mse - mse.min()) / (mse.max() - mse.min())  # Normaliser les MSE pour les utiliser comme pseudo-probabilités
fraction_of_positives, mean_predicted_value = calibration_curve(Y_test, mse_normalized, n_bins=10)

# Calculer les courbes de calibration pour le modèle avec SMOTE
mse_smote_normalized = (mse_smote - mse_smote.min()) / (mse_smote.max() - mse_smote.min())  # Normaliser
fraction_of_positives_smote, mean_predicted_value_smote = calibration_curve(Y_test, mse_smote_normalized, n_bins=10)

# Tracer les courbes de calibration des deux modèles
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Modèle sans SMOTE
axes[0].plot(mean_predicted_value, fraction_of_positives, marker='o', label='Autoencoder sans SMOTE')
axes[0].plot([0, 1], [0, 1], linestyle='--', label='Perfectly calibrated')
axes[0].set_title("Courbe de Calibration (Sans SMOTE)")
axes[0].set_xlabel("Probabilité Moyenne Prédite")
axes[0].set_ylabel("Fraction d'Observations Positives")
axes[0].legend()

# Modèle avec SMOTE
axes[1].plot(mean_predicted_value_smote, fraction_of_positives_smote, marker='o', label='Autoencoder avec SMOTE')
axes[1].plot([0, 1], [0, 1], linestyle='--', label='Perfectly calibrated')
axes[1].set_title("Courbe de Calibration (Avec SMOTE)")
axes[1].set_xlabel("Probabilité Moyenne Prédite")
axes[1].set_ylabel("Fraction d'Observations Positives")
axes[1].legend()

# Afficher les graphes
plt.tight_layout()
plt.show()
```
