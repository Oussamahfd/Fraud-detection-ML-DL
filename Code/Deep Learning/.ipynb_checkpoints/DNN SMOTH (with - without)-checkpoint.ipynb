```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, f1_score, classification_report
```

# <span style="color:black;">Préparation et Prétraitement des Données pour la Détection de Fraude
</span>


## <span style="color:black;">Chargement des datasets</span>


```python
# Chargement des données
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

## <span style="color:black;">Suppression des colonnes inutiles</span>


```python
# Prétraitement des données

# Supprimer les colonnes non pertinentes
df_train = df_train.drop(['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'trans_num'], axis=1)
df_test = df_test.drop(['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'trans_num'], axis=1)

# Convertir les variables catégorielles en variables numériques
categorical_cols = df_train.select_dtypes(include=['object']).columns
df_train[categorical_cols] = df_train[categorical_cols].apply(lambda x: LabelEncoder().fit_transform(x))
df_test[categorical_cols] = df_test[categorical_cols].apply(lambda x: LabelEncoder().fit_transform(x))
```

## <span style="color:black;">Séparation des features et de la target</span>


```python
# Séparer les features et la cible
X_train = df_train.drop('is_fraud', axis=1)
y_train = df_train['is_fraud']
X_test = df_test.drop('is_fraud', axis=1)
y_test = df_test['is_fraud']
```

```python
# Normaliser les données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# <span style="color:black;">Entraînement du Modèle de DNN [Sans / Avec] SMOTE</span>


## <span style="color:black;">Sans SMOTE</span>

```python
# Définition de l'architecture du modèle DNN
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),  # Pour réduire le sur-apprentissage
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # Sortie binaire
])
```

#### <span style="color:black;">Évaluation des Performances du Modèle de DNN</span>


```python
# Compilation et entraînement du modèle
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Prédictions
y_pred_proba = model.predict(X_test).flatten()  # Probabilités prédites

# Calculer les courbes ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Afficher les courbes ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Afficher la matrice de confusion
y_pred = (y_pred_proba > 0.5).astype(int)  # Convertir les probabilités en classes binaires
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Rapport de classification
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

## <span style="color:black;">Avec SMOTE</span>

```python
# Appliquer SMOTE sur les données d'entraînement
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Définition de l'architecture du modèle DNN
model_smote = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_smote.shape[1],)),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

#### <span style="color:black;">Évaluation des Performances du Modèle de DNN avec SMOTE</span>


```python
# 5. Compilation et entraînement du modèle avec SMOTE
model_smote.compile(optimizer=Adam(learning_rate=0.001),
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

history_smote = model_smote.fit(X_train_smote, y_train_smote, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Prédictions
y_pred_proba_smote = model_smote.predict(X_test).flatten()
```

## <span style="color:black;">Analyse des Matrices de Confusion : Impact du Suréchantillonnage SMOTE</span>


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Convertir les probabilités prédites en classes (0 ou 1)
y_pred = (y_pred_proba > 0.5).astype(int)  # Modèle sans SMOTE
y_pred_smote = (y_pred_proba_smote > 0.5).astype(int)  # Modèle avec SMOTE

# Calcul des matrices de confusion
cm = confusion_matrix(y_test, y_pred)
cm_smote = confusion_matrix(y_test, y_pred_smote)

# Affichage côte à côte des matrices de confusion
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Modèle sans SMOTE
ConfusionMatrixDisplay(cm).plot(ax=axes[0], cmap='Blues')
axes[0].set_title("Confusion Matrix Without SMOTE")

# Modèle avec SMOTE
ConfusionMatrixDisplay(cm_smote).plot(ax=axes[1], cmap='Blues')
axes[1].set_title("Confusion Matrix With SMOTE")

plt.tight_layout()
plt.show()
```



### Analyse des graphiques
Les matrices de confusion pour les modèles LSTM avec et sans SMOTE montrent les résultats suivants :
- **Modèle sans SMOTE** :
  - Vrais négatifs : 552873 (bon pour identifier les exemples négatifs).
  - Faux négatifs : 1095 (relativement faible, peu d'exemples positifs manqués).
  - Faux positifs : 701 (non négligeable selon le contexte).
  
- **Modèle avec SMOTE** :
  - Vrais négatifs : 534921 (légère diminution, reste élevé).
  - Faux négatifs : 372 (amélioration par rapport au modèle sans SMOTE).
  - Faux positifs : 18653 (augmentation significative, potentiellement problématique).

### Interprétation des résultats
- **Impact de SMOTE** : SMOTE a réduit le nombre de faux négatifs, augmentant ainsi la capacité à identifier les exemples positifs, mais a entraîné une augmentation des faux positifs.
- **Déséquilibre des classes** : Le déséquilibre entre les classes explique les résultats, SMOTE a amélioré le rappel au détriment de la précision.

### Conclusion
L'analyse des matrices de confusion révèle que SMOTE a amélioré le rappel, mais au coût de la précision. Le choix entre les modèles dépendra du coût relatif des faux positifs et des faux négatifs dans le contexte d'application.


## <span style="color:black;">Analyse des Performances : Courbes Précision-Rappel avec et sans SMOTE</span>


```python
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay

# Calcul des courbes précision-rappel pour le modèle sans SMOTE
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)

# Calcul des courbes précision-rappel pour le modèle avec SMOTE
precision_smote, recall_smote, _ = precision_recall_curve(y_test, y_pred_proba_smote)

# Création de la figure pour les deux graphiques côte à côte
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Courbe précision-rappel pour le modèle sans SMOTE
PrecisionRecallDisplay(precision=precision, recall=recall).plot(ax=axes[0])
axes[0].set_title("Precision-Recall Curve Without SMOTE")

# Courbe précision-rappel pour le modèle avec SMOTE
PrecisionRecallDisplay(precision=precision_smote, recall=recall_smote).plot(ax=axes[1])
axes[1].set_title("Precision-Recall Curve With SMOTE")

plt.tight_layout()
plt.show()
```

### Compréhension des courbes de précision-rappel
Les courbes de précision-rappel sont cruciales pour évaluer les modèles de classification binaire, en particulier avec des classes déséquilibrées. Elles montrent le compromis entre :
- **Précision** : Proportion d'exemples positifs correctement prédits.
- **Rappel** : Proportion d'exemples positifs correctement identifiés.

### Analyse des graphiques
Les graphiques montrent les courbes de précision-rappel pour un modèle LSTM, avec et sans SMOTE :
- **Modèle sans SMOTE** : La courbe montre une décroissance rapide de la précision à mesure que le rappel augmente, indiquant une augmentation des faux positifs.
- **Modèle avec SMOTE** : La courbe est plus élevée, surtout pour des valeurs de rappel élevées, permettant d'identifier davantage d'exemples positifs tout en maintenant une précision raisonnable.

### Interprétation des résultats
- **Impact de SMOTE** : L'application de SMOTE a significativement amélioré le rappel, rendant le modèle plus efficace pour identifier les exemples positifs, essentiel lorsque les faux négatifs sont coûteux.
- **Compromis précision-rappel** : L'amélioration du rappel se fait souvent au détriment de la précision, soulignant le compromis classique en classification binaire.
- **Choix du seuil** : Le seuil de classification doit être choisi en fonction de l'application. Un seuil plus bas minimisera les faux négatifs, tandis qu'un seuil plus élevé minimisera les faux positifs.

### Conclusion
L'analyse des courbes de précision-rappel indique que l'utilisation de SMOTE a amélioré le rappel du modèle LSTM tout en maintenant une précision raisonnable. Cela suggère que SMOTE est efficace pour traiter les déséquilibres de classes dans les problèmes de classification.


## <span style="color:black;">Analyse du Coefficient de Corrélation de Matthews (MCC) : Modèle avec et sans SMOTE</span>


```python
from sklearn.metrics import matthews_corrcoef

# Convertir les probabilités prédites en classes binaires
y_pred = (y_pred_proba >= 0.5).astype(int)
y_pred_smote = (y_pred_proba_smote >= 0.5).astype(int)

# Calculer le MCC pour le modèle sans SMOTE
mcc = matthews_corrcoef(y_test, y_pred)

# Calculer le MCC pour le modèle avec SMOTE
mcc_smote = matthews_corrcoef(y_test, y_pred_smote)

# Afficher les MCC dans un graphique
labels = ['Without SMOTE', 'With SMOTE']
mcc_values = [mcc, mcc_smote]

# Création de la figure avec deux barres côte à côte
fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(labels, mcc_values, color=['blue', 'green'])
ax.set_title("Matthews Correlation Coefficient (MCC) Comparison")
ax.set_ylabel("MCC Value")

# Afficher les valeurs de MCC sur les barres
for i, v in enumerate(mcc_values):
    ax.text(i, v + 0.01, f'{v:.2f}', ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```

### Analyse du coefficient de corrélation de Matthews (MCC)

#### Compréhension du coefficient de corrélation de Matthews (MCC)
Le MCC est une mesure de la qualité des classifications binaires, prenant en compte les quatre valeurs d'une matrice de confusion (vrais positifs, vrais négatifs, faux positifs et faux négatifs). Il fournit un score entre -1 et 1 :
- **1** : Prédiction parfaite
- **-1** : Prédiction totalement inverse
- **0** : Prédiction aléatoire

#### Analyse du graphique
- **Modèle sans SMOTE** : MCC de **0.54**, indiquant une corrélation positive modérée et une capacité à distinguer les classes.
- **Modèle avec SMOTE** : MCC de **0.26**, signalant une dégradation des performances et une capacité réduite à discriminer les classes.


#### Conclusions
- **SMOTE n'est pas toujours bénéfique** : Son application ne garantit pas une amélioration des performances et peut parfois les dégrader.
- **Choix de la technique d'équilibrage** : Il est crucial de sélectionner la méthode d'équilibrage la plus adaptée au problème et au jeu de données.
- **Importance de l'évaluation** : Évaluer les performances du modèle avec plusieurs métriques (précision, rappel, F1-score, MCC) est essentiel pour une vision complète de ses capacités.
