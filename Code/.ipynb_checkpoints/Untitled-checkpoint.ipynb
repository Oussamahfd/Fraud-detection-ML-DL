```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report ,f1_score
from imblearn.over_sampling import SMOTE
```

```python
df_train = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTrain.csv')
df_test = pd.read_csv('C:/Users/PC/Desktop/2/pfa/Datasets/Credit Card Transactions Fraud Detection Dataset/fraudTest.csv')
```

```python
df_train.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
df_test.drop(columns=['Unnamed: 0', 'merchant', 'cc_num', 'first', 'last', 'gender', 'trans_num', 'unix_time', 'street', 'merch_lat', 'merch_long', 'job', 'zip'], inplace=True)
```

```python
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_train['dob'] = pd.to_datetime(df_train['dob']).apply(lambda x: x.timestamp())

df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time']).apply(lambda x: x.timestamp())
df_test['dob'] = pd.to_datetime(df_test['dob']).apply(lambda x: x.timestamp())
```

```python
label_encoder = LabelEncoder()
columns_to_convert_in_string_to_int = ['category', 'city', 'state']
for col in columns_to_convert_in_string_to_int:
    df_train[col] = label_encoder.fit_transform(df_train[col])
    df_test[col] = label_encoder.fit_transform(df_test[col])
```

```python
columns_to_convert_in_int_to_float = ['category', 'city', 'state', 'city_pop']
for col in columns_to_convert_in_int_to_float:
    df_train[col] = df_train[col].astype('float64')
    df_test[col] = df_test[col].astype('float64')
```

```python
X = df_train.drop(columns=['is_fraud'])
Y = df_train['is_fraud']
```

```python
# Variables indépendantes (features) et dépendantes (label) pour l'entraînement
X_train = df_train.drop(columns=['is_fraud'])  # Supprimer la colonne de la cible
Y_train = df_train['is_fraud']  # La colonne cible 'is_fraud'

# Variables indépendantes (features) et dépendantes (label) pour le test
X_test = df_test.drop(columns=['is_fraud'])  # Supprimer la colonne de la cible
Y_test = df_test['is_fraud']  # La colonne cible 'is_fraud'

# Standardisation des données
scaler = StandardScaler()

# Appliquer la normalisation uniquement sur les variables indépendantes
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Affichage des formes des datasets obtenus
print("Shape of X_train:", X_train.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
# Importation des bibliothèques nécessaires
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, f1_score

# Création et entraînement du modèle de régression logistique sans SMOTE
model_without_smote = LogisticRegression(max_iter=1000, random_state=42)
model_without_smote.fit(X_train, Y_train)

# Prédictions sur les données d'entraînement
y_train_pred_without_smote = model_without_smote.predict(X_train)

# Prédictions sur les données de test
y_test_pred_without_smote = model_without_smote.predict(X_test)

# Évaluation des performances du modèle
train_accuracy = accuracy_score(Y_train, y_train_pred_without_smote)
test_accuracy = accuracy_score(Y_test, y_test_pred_without_smote)

# Affichage des résultats
print("Accuracy on training data (without SMOTE):", train_accuracy)
print("Accuracy on test data (without SMOTE):", test_accuracy)
print("Classification report (Test Data):\n", classification_report(Y_test, y_test_pred_without_smote))
```

```python
# Importation de SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Application de SMOTE pour équilibrer les classes
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
# Affichage des formes des datasets obtenus
print("Shape of X_train:", X_train_smote.shape)
print("Shape of Y_train:", Y_train_smote.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python

# Création et entraînement du modèle de régression logistique avec SMOTE
model_with_smote = LogisticRegression(max_iter=1000, random_state=42)
model_with_smote.fit(X_train_smote, Y_train_smote)

# Prédictions sur les données d'entraînement et de test
y_train_pred_with_smote = model_with_smote.predict(X_train_smote)
y_test_pred_with_smote = model_with_smote.predict(X_test)

# Évaluation des performances du modèle
train_accuracy_with_smote = accuracy_score(Y_train_smote, y_train_pred_with_smote)
test_accuracy_with_smote = accuracy_score(Y_test, y_test_pred_with_smote)

# Affichage des résultats
print("Accuracy on training data (with SMOTE):", train_accuracy_with_smote)
print("Accuracy on test data (with SMOTE):", test_accuracy_with_smote)
print("Classification report (Test Data, with SMOTE):\n", classification_report(Y_test, y_test_pred_with_smote))
```

```python
import numpy as np

# Fonction sigmoïde
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Fonction de coût (log-loss)
def compute_cost(X, Y, weights, bias):
    m = X.shape[0]
    Z = np.dot(X, weights) + bias
    A = sigmoid(Z)
    cost = -1/m * np.sum(Y * np.log(A) + (1 - Y) * np.log(1 - A))
    return cost

# Descente de gradient pour mettre à jour les poids et le biais
def gradient_descent(X, Y, weights, bias, learning_rate, num_iterations):
    m = X.shape[0]
    
    for i in range(num_iterations):
        # Prédictions
        Z = np.dot(X, weights) + bias
        A = sigmoid(Z)
        
        # Calcul des gradients
        dw = 1/m * np.dot(X.T, (A - Y))
        db = 1/m * np.sum(A - Y)
        
        # Mise à jour des poids et du biais
        weights -= learning_rate * dw
        bias -= learning_rate * db
        
        # Calcul du coût (facultatif : pour suivre la progression)
        if i % 100 == 0:
            cost = compute_cost(X, Y, weights, bias)
            print(f"Cost after iteration {i}: {cost}")
    
    return weights, bias

# Modèle de régression logistique
class LogisticRegressionModel:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, Y):
        # Initialiser les paramètres
        n_features = X.shape[1]
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Appliquer la descente de gradient
        self.weights, self.bias = gradient_descent(X, Y, self.weights, self.bias, self.learning_rate, self.num_iterations)
    
    def predict(self, X):
        # Prédictions avec la fonction sigmoïde
        Z = np.dot(X, self.weights) + self.bias
        A = sigmoid(Z)
        return (A >= 0.5).astype(int)  # Seuil de 0.5 pour classifier


```

```python
# Utilisation du modèle
# Supposons que X_train et Y_train soient déjà normalisés et prêts à l'emploi
model = LogisticRegressionModel(learning_rate=0.01, num_iterations=1000)
model.fit(X_train, Y_train)

# Prédictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Évaluation des performances
train_accuracy = accuracy_score(Y_train, y_train_pred)
test_accuracy = accuracy_score(Y_test, y_test_pred)
print(f"Accuracy on training data: {train_accuracy}")
print(f"Accuracy on test data: {test_accuracy}")
```
